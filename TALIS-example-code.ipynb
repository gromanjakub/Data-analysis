```python
import pandas as pd
import numpy as np
import pyreadstat
df_cesi=pd.read_spss("C:/Users/Jakub/Downloads/talis/cz/PTGCZET3-----.sav") 
df_intvsichni=pd.read_spss("C:/Users/Jakub/Downloads/talis/PTGINTT3.sav") 

from matplotlib import pyplot as plt
from matplotlib import *
roky = {0,1,2}
df_cesi_stari = df_cesi[~df_cesi['TT3G11B'].isin(roky)]
df_cesi_mladi = df_cesi[df_cesi['TT3G11B'].isin(roky)]
df_int_mladi = df_intvsichni[df_intvsichni['TT3G11B'].isin(roky)]
```

```python
df_cesi_mladi = df_cesi_mladi[df_cesi_mladi['TT3G27A'].notna()]
df_cesi_stari = df_cesi_stari[df_cesi_stari['TT3G27A'].notna()]
df_int_mladi = df_int_mladi[df_int_mladi['TT3G27A'].notna()]
```

```python
data_18= [
     df_cesi_mladi['TT3G27A'], df_cesi_stari['TT3G27A'],
    df_int_mladi['TT3G27A']]
```

```python
df_cesi_mladi['TT3G27A'].value_counts(normalize=True)
```

```python
#27a
labels18 = ["No need", "Low need", "Moderate need", "High need"]
x18 = np.arange(len(labels18))
fig18 = plt.figure(figsize=(18,10))
ax18 = fig18.add_axes([0,0,1,1])

ax18.set_xticks(x18)
ax18.set_xticklabels(labels18)

font = {#'family' : 'normal',
        'weight' : 'bold',
        'size'   : 26}

plt.rc('font', **font)

ax18.bar(x18, df_cesi_mladi['TT3G27A'].value_counts(normalize=True), color ="b", width = 0.25)
ax18.bar(x18+0.25, df_cesi_stari['TT3G27A'].value_counts(normalize=True), color ="g", width = 0.25)
ax18.bar(x18+0.5, df_int_mladi['TT3G27A'].value_counts(normalize=True), color ="r", width = 0.25)

def add_value_labels(ax, typ, spacing=5):
    space = spacing
    va = 'bottom'

    for i in ax18.patches:
                y_value = i.get_height()
                x_value = i.get_x() + i.get_width() / 2.0
                label = "{:.2f}".format(y_value)
                ax18.annotate(label,(x_value, y_value), xytext=(0.0, space), 
                        textcoords="offset points", ha='center', va=va, rotation = 90)  

add_value_labels(ax18, typ='bar')
```
